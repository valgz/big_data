```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra

def precio_total(articulos, compras):
    precios = list(map(lambda item: articulos.get(item), compras))
    return reduce(lambda prev, actual: prev + actual, precios)

# B. Una función que calcule el precio total de la compra con un descuento del 10%

def descuento_10(articulos, compras):
    total = precio_total(articulos, compras)
    descuento = total * 0.1
    return total - descuento

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def mayor_a_100(articulos, compras):
    if precio_total(articulos, compras) > 100:
        return descuento_10(articulos, compras)
    else:
        return precio_total(articulos, compras)
        
# D. Una función que calcule el precio total aplicando el IVA (21%)

def aplica_iva(articulos, compras):
    total = precio_total(articulos, compras)
    impuesto = total * 0.21
    return total + impuesto

# E.lista los artículos cuyo precio es superior a 20€

def superior_a_70(articulos):
    return list(filter(lambda articulo: articulos.get(articulo) > 20, articulos))
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]

from functools import reduce

def boletin_por_grado(curso, alumnos=alumnos):
    
    if curso == 1 or curso == 2:
        seleccion = list(filter(lambda alumno: alumno[2] == str(curso), alumnos))
        
        lista_alumnos = list(map(lambda alumno: list(alumno), seleccion))
        
        promedios = list(map(lambda alumno: reduce(lambda prev, actual: prev + actual, alumno[3]) / len(alumno[3]), lista_alumnos))

        nombres = list(map(lambda alumnos: alumnos[0] + ' ' + alumnos[1], lista_alumnos))

        resultados = list(map(lambda alumno: f'{alumno[0]}. Promedio: {alumno[1]}', zip(nombres, promedios)))

        return resultados
    else:
        return 'El valor introducido no es válido. Por favor, ingrese 1 o 2'
            
```
